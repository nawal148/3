```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('google_books_dataset.csv')
#    
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-06', '2023-01-08', '2023-01-09', '2023-01-10'] * 10,
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Food', 'Clothing', 'Food', 'Food', 'Electronics', 'Clothing', 'Food'] * 10,
    'Sales': [1500, 200, 1500, 50, 300, 60, 60, 10000, 250, 70] * 10,
    'Quantity': [10, 5, np.nan, 20, 2, 30, 30, 1, 4, 15] * 10,
    'CustomerID': [1, 2, 3, 4, 5, 6, 6, 8, 9, 10] * 10
}
df = pd.DataFrame(data)
df['Date'] = df['Date'].astype(str)

print(df.head())
print(df.tail())
df.info()
df['Date'] = pd.to_datetime(df['Date'])
print(df.dtypes)

print(df.isnull().sum())
median_qty = df['Quantity'].median()
df['Quantity'].fillna(median_qty, inplace=True)
print(df.isnull().sum())

print(df.duplicated().sum())
df.drop_duplicates(inplace=True)

sns.boxplot(x=df['Sales'])
plt.show()
df['z_score'] = (df['Sales'] - df['Sales'].mean()) / df['Sales'].std()
print(df[df['z_score'].abs() > 3][['Sales', 'z_score']])
df.drop('z_score', axis=1, inplace=True)

print(df.describe())
print(df['Category'].value_counts())
print(df.corr(numeric_only=True))

sns.histplot(df['Sales'], kde=True)
plt.show()
daily_sales = df.groupby('Date')['Sales'].sum().reset_index()
sns.lineplot(data=daily_sales, x='Date', y='Sales')
plt.show()
sns.countplot(data=df, x='Category')
plt.show()
sns.boxplot(data=df, x='Category', y='Sales')
plt.show()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()

print(df.groupby('Category')['Sales'].mean())
print(df.groupby('Category').agg({'Quantity': 'sum', 'Sales': 'sum'}))
print(df.pivot_table(values='Sales', index='Date', columns='Category', aggfunc='sum'))
```
